```python
import nltk
from nltk import word_tokenize, sent_tokenize
import gensim
from gensim.models.word2vec import Word2Vec
from sklearn.manifold import TSNE
import pandas as pd
from bokeh.io import output_notebook
from bokeh.plotting import show, figure
```

```python
nltk.download('punkt')
```

```python
nltk.download('gutenberg')
```

```python
from nltk.corpus import gutenberg
print(gutenberg.fileids())
```

```python
text = gutenberg.raw()
tokenz=word_tokenize(text)
print(len(tokenz))
```

```python
#  getting unique words
uniq=[]
for w in tokenz:
  if w not  in uniq:
    uniq.append(w)
print(len(uniq))
print(uniq)
```

```python
uniq_lower=[w.lower() for w in uniq]
print(len(uniq_lower))
print(uniq_lower)
```

```python
guten_sentece=gutenberg.sents()
print(guten_sentece)
```

```python
#  word to vec to create a word vector
model = Word2Vec(sentences=guten_sentece, vector_size=64, sg=1, window=10, min_count=5, workers=8)
model.save("raw_gutenberg_model.w2v")

# Load the model (for future use)
model = gensim.models.Word2Vec.load('raw_gutenberg_model.w2v')
```
